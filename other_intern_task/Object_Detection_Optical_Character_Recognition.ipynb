```python
#Importing The Required Libraries

import cv2
import matplotlib.pyplot as plt
import easyocr
from pylab import rcParams
```

```python
from IPython.display import Image
rcParams['figure.figsize'] = 10,20
```


# **Downloading English language model**

```python
Lang_reader = easyocr.Reader(['en'])
```

# **Image preview**

```python
Image('/content/drive/MyDrive/images/love.jpg')
```

# **Detection of texts in Love.jpeg**

```python
output = Lang_reader.readtext('/content/drive/MyDrive/images/love.jpg')
```

# **Print the detected information from texts in image**
This list shows bounding box co-ordinates, Text and Confidence level

```python
print(output)
```

# **Extract the fetched output text & bounding box coordinates**

```python
cord1 = output[-4][0]
print(output[-4][1])
```

# **Minimum value of coordinates**

```python
x_min1,y_min1 = [min(idx) for idx in zip(*cord1)]
```

# **Maximum value of coordinates**

```python
x_max1,y_max1 = [max(idx) for idx in zip(*cord1)]
```

# **Visualization of bounding box co-ordinates of particular text**

```python
image = cv2.imread('/content/drive/MyDrive/images/love.jpg')
cv2.rectangle(image,(int(x_min1),int(y_min1)),(int(x_max1),int(y_max1)),(0,255,0),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
